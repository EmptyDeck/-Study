```python
import pandas as pd


# Concatenate the two DataFrames along the row axis
df = pd.concat([x_train, x_test], ignore_index=True)
#remove enrollee id (since its irrelivent)
df = df.drop(['enrollee_id'], axis=1)
print(df.head(3))
```

## NAN replace (with most frequent)

```python
# Check for NaN values in each label
nan_labels = df.columns[df.isna().any()].tolist()

print(nan_labels)
```

```python
columns_with_nan = ['gender', 'enrolled_university', 'education_level', 'major_discipline', 'experience', 'company_size', 'company_type', 'last_new_job']

for col in columns_with_nan:
    df[col].fillna(df[col].mode()[0], inplace=True)
print(df.head(3))
```

```python
# RE Check for NaN values in each label
nan_labels = df.columns[df.isna().any()].tolist()

print(nan_labels)
```

## Encoding to int

```python
#str to int
from sklearn.preprocessing import LabelEncoder

# Initialize a LabelEncoder
le = LabelEncoder()

# List of columns to encode
cols_to_encode = ['city', 'gender', 'relevent_experience', 'enrolled_university', 
                    'education_level', 'major_discipline', 'company_type']

# Iterate over each categorical column and transform it
for col in cols_to_encode:
    df[col] = le.fit_transform(df[col])
    
print(df.head(3))
```

## Encoding 2

```python
cols_to_handle = ['experience', 'company_size', 'last_new_job']

for col in cols_to_handle:
    print(f"Unique values for {col}: {df[col].unique()}")
```

```python
def handle_experience(x):
    if pd.isnull(x):
        return x  # or define a default value
    elif x == '>20':
        return 23
    elif x == '<1':
        return 0
    else:
        return int(x)

def handle_company_size(x):
    if pd.isnull(x):
        return x  # or define a default value
    elif x == '<10':
        return 5
    elif x == '10/49':
        return 30
    elif x == '50-99':
        return 75
    elif x == '100-500':
        return 300
    elif x == '500-999':
        return 750
    elif x == '1000-4999':
        return 3000
    elif x == '5000-9999':
        return 7500
    elif x == '10000+':
        return 15000

def handle_last_new_job(x):
    if pd.isnull(x):
        return x  # or define a default value
    elif x == 'never':
        return 0
    elif x == '>4':
        return 5
    else:
        return int(x)

# Applying the transformations
df['experience'] = df['experience'].map(handle_experience)
df['company_size'] = df['company_size'].map(handle_company_size)
df['last_new_job'] = df['last_new_job'].map(handle_last_new_job)

print(df.head(3))
```

## Back to train, test

```python
import pandas as pd

# Split the concatenated DataFrame back into x_train and x_test
x_train = df.loc[:len(x_train)-1, :]
x_test = df.loc[len(x_train):, :]

# Reset the index for each DataFrame
x_train.reset_index(drop=True, inplace=True)
x_test.reset_index(drop=True, inplace=True)
```

## valdation test / check acc

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the original training data into training and validation set
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

clf = QuadraticDiscriminantAnalysis()
clf.fit(x_train,y_train)

# Predict on validation data
y_val_pred = clf.predict(x_val)

# Print the accuracy score
print("Accuracy on validation data: %.4f "%accuracy_score(y_val, y_val_pred))
```
