```python
#classification tree

import seaborn as sns
iris = sns.load_dataset('iris')
x = iris.drop('species' , axis = 1)
y = iris['species']
```

```python
from sklearn.preprocessing import LabelEncoder
classes= LabelEncoder()
y = classes.fit_transform(iris['species'].values)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,stratify = y)
```

```python
from sklearn import tree

dtc = tree.DecisionTreeClassifier(criterion='entropy' , max_depth=7,random_state=1)

dtc.fit(xtrain,ytrain)

y_train_pred = dtc.predict(xtrain)
y_test_pred = dtc.predict(xtest)

```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytrain,y_train_pred))
print(accuracy_score(ytest,y_test_pred)) 
```

```python
import matplotlib.pyplot as plt

fig , ax = plt.subplots(figsize = (25,12))
tree.plot_tree(dtc.fit(xtrain,ytrain),fontsize=15)
plt.savefig('tree_high_dpi',dpi = 100)
```
