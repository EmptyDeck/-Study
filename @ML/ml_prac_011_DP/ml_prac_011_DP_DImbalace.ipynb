```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE,ADASYN

x,y = make_classification(n_classes=3,weights=[0.03,0.07,0.9],n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,n_samples=200,random_state=10)

print('Origianl dataset shape %s' %Counter(y))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[:,0],x[:,1],marker = 'o' ,c=y,s=100,edgecolors="k",linewidths=1)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

# Data Imbalance fix

# SMOTE

```python
sm = SMOTE()
xres,yres = sm.fit_resample(x,y)
print('resampled : %s'%Counter(yres))
```

```python
import matplotlib.pyplot as plt

plt.scatter(xres[:,0],xres[:,1],marker = 'o' ,c=yres,s=100,edgecolors="k",linewidths=1)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

# ADASYN

```python
ada = ADASYN()
xsyn , ysyn = ada.fit_resample(x,y)
print('Resampled : %s'%Counter(ysyn))
```

```python
import matplotlib.pyplot as plt

plt.scatter(xsyn[:,0],xsyn[:,1],marker = 'o' ,c=ysyn,s=100,edgecolors="k",linewidths=1)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```

# NearMiss

```python
from imblearn.under_sampling import NearMiss

under = NearMiss(version=3,n_neighbors_ver3=3)
xunder,yunder = under.fit_resample(x,y)
print('Resampled : %s'%Counter(yunder))
```

```python
import matplotlib.pyplot as plt

plt.scatter(xunder[:,0],xunder[:,1],marker = 'o' ,c=yunder,s=100,edgecolors="k",linewidths=1)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()
```
