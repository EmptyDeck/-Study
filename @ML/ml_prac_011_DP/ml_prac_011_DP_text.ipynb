```python
text = ['what up everyone',
        'how are you guys doing? is every thing okay for all of you? I hope you have a great day',
        'did you take that out? you didnt? why not? where you busy?']
text
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec2 = CountVectorizer()
t = vec2.fit_transform(text).toarray()

import pandas as pd
t1 = pd.DataFrame(t, columns=vec2.get_feature_names_out())
t1
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfid = TfidfTransformer()
x2 = tfid.fit_transform(t1).toarray()
x3 = pd.DataFrame(x2,columns=tfid.get_feature_names_out())
x3
```
