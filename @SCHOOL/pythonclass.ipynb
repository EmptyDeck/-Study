```python
import tkinter as tk
from tkinter import messagebox

class Calculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Tkinter Calculator")
        
        # 계산기의 디스플레이
        self.display = tk.Entry(self.root, font=("Arial", 20), borderwidth=5, relief="ridge", justify="right")
        self.display.grid(row=0, column=0, columnspan=4, pady=10, padx=10, sticky="we")

        # 계산기 상태
        self.equation = ""  # 현재 수식을 저장합니다.
        
        # 숫자 및 연산자 버튼을 생성합니다.
        self.create_buttons()

    def create_buttons(self):
        # 버튼에 사용할 텍스트와 위치를 매핑하는 리스트를 생성합니다.
        buttons = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
            ('C', 5, 0), ('%', 5, 1), ('√', 5, 2), ('^2', 5, 3)
        ]

        for (text, row, col) in buttons:
            button = tk.Button(self.root, text=text, font=("Arial", 20), width=5, height=2,
                               command=lambda btn=text: self.on_button_click(btn))
            button.grid(row=row, column=col, sticky="nsew", padx=5, pady=5)

        # 그리드 설정: 모든 행과 열에 동일한 공간을 할당합니다.
        for i in range(1, 6):
            self.root.grid_rowconfigure(i, weight=1)
        for i in range(4):
            self.root.grid_columnconfigure(i, weight=1)

    def on_button_click(self, char):
        # 버튼 클릭 이벤트를 처리합니다.
        if char == '=':
            self.calculate_result()
        elif char == 'C':
            self.clear_display()
        elif char == '√':
            self.calculate_sqrt()
        elif char == '^2':
            self.calculate_square()
        elif char == '%':
            self.calculate_percentage()
        else:
            self.append_to_equation(char)

    def calculate_result(self):
        # 수식의 결과를 계산하고 표시합니다.
        try:
            result = str(eval(self.equation))
            self.clear_display()
            self.display.insert(0, result)
        except Exception as e:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_sqrt(self):
        # 제곱근을 계산합니다.
        try:
            result = str(eval(self.equation) ** 0.5)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_square(self):
        # 제곱을 계산합니다.
        try:
            result = str(eval(self.equation) ** 2)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_percentage(self):
        # 퍼센트를 계산합니다.
        try:
            result = str(eval(self.equation) * 0.01)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def clear_display(self):
        # 디스플레이를 클리어합니다.
        self.display.delete(0, tk.END)
        self.equation = ""

    def append_to_equation(self, char):
        # 디스플레이에 문자를 추가합니다.
        self.equation += str(char)
        self.display.delete(0, tk.END)
        self.display.insert(tk.END, self.equation)

# Tkinter 윈도우를 생성하고, 계산기 앱을 시작합니다.
if __name__ == "__main__":
    root = tk.Tk()
    app = Calculator(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import math

class Calculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Tkinter Calculator")
        
        # Calculator display
        self.display = tk.Entry(self.root, font=("Arial", 20), borderwidth=5, relief="ridge", justify="right")
        self.display.grid(row=0, column=0, columnspan=4, pady=10, padx=10, sticky="we")

        # Calculator state
        self.equation = ""  # Stores the current equation.
        
        # Create buttons for the calculator
        self.create_buttons()

    def create_buttons(self):
        # Create a list that maps the text and position of buttons
        buttons = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
            ('C', 5, 0), ('%', 5, 1), ('√', 5, 2), ('^2', 5, 3),
            ('sin', 6, 0), ('cos', 6, 1), ('tan', 6, 2), ('π', 6, 3)
        ]

        for (text, row, col) in buttons:
            button = tk.Button(self.root, text=text, font=("Arial", 20), width=5, height=2,
                               command=lambda btn=text: self.on_button_click(btn))
            button.grid(row=row, column=col, sticky="nsew", padx=5, pady=5)

        # Grid configuration: Allocate equal space to all rows and columns.
        for i in range(1, 7):
            self.root.grid_rowconfigure(i, weight=1)
        for i in range(4):
            self.root.grid_columnconfigure(i, weight=1)

    def on_button_click(self, char):
        # Handle button click events.
        if char == '=':
            self.calculate_result()
        elif char == 'C':
            self.clear_display()
        elif char == '√':
            self.calculate_sqrt()
        elif char == '^2':
            self.calculate_square()
        elif char == '%':
            self.calculate_percentage()
        elif char in ['sin', 'cos', 'tan']:
            self.calculate_trigonometry(char)
        elif char == 'π':
            self.append_to_equation('*' + str(math.pi))
        else:
            self.append_to_equation(char)

    def calculate_result(self):
        # Calculate and display the result of the equation.
        try:
            result = str(eval(self.equation))
            self.clear_display()
            self.display.insert(0, result)
        except Exception as e:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_trigonometry(self, trig_function):
        # Calculate sine, cosine, or tangent of the given value.
        try:
            # Extract the value from the display for trigonometric calculation.
            value = float(self.display.get())
            if trig_function == 'sin':
                result = str(math.sin(math.radians(value)))
            elif trig_function == 'cos':
                result = str(math.cos(math.radians(value)))
            elif trig_function == 'tan':
                result = str(math.tan(math.radians(value)))
            
            self.clear_display()
            self.display.insert(0, result)
        except Exception as e:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_sqrt(self):
        # Calculate the square root.
        try:
            result = str(eval(self.equation) ** 0.5)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_square(self):
        # Calculate the square.
        try:
            result = str(eval(self.equation) ** 2)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def calculate_percentage(self):
        # Calculate the percentage.
        try:
            result = str(eval(self.equation) * 0.01)
            self.clear_display()
            self.display.insert(0, result)
        except:
            messagebox.showerror("Error", "Invalid Input")
            self.clear_display()

    def append_to_equation(self, char):
        # Append a character to the display.
        self.equation += str(char)
        self.display.delete(0, tk.END)
        self.display.insert(tk.END, self.equation)

    def clear_display(self):
        # Clear the display.
        self.display.delete(0, tk.END)
        self.equation = ""

# Create the Tkinter window and start the calculator app.
if __name__ == "__main__":
    root = tk.Tk()
    app = Calculator(root)
    root.mainloop()
```
