```python
import pandas as pd

base_stock_data=pd.read_excel('data/NaverFinance.xlsx', sheet_name='Sheet1', thousands = ',')

# 1. 불필요한 컬럼을 없앤다
# (액면가, 거래대금, 매수호가, 매도호가, 외국인 비율 등등)
def getTrimData(base_stock_data) :
    trim_data = base_stock_data[['종목명','거래량','상장주식수','등락률','현재가','시가','고가','PER','시가총액','매출액','PBR']]
    return trim_data

# 2. 우량한 주식을 선정한다.
# 시가총액 1000억 이상, 매출액 1억 이상, per 0 이상, pbr 2이하
# 이들 중 Null인 값이 하나라도 있으면 해당 주식은 제외
# PBR의 역수가 큰 순서대로 100개의 주식을 우량주식으로 선정함.
def getBlueChip100(trim_data):
    trim_data = trim_data.loc[trim_data['시가총액'].notna() & trim_data['매출액'].notna()& trim_data['PER'].notna()& trim_data['PBR'].notna()]
    trim_data = trim_data.loc[
        (trim_data['시가총액'] >= 1000) & 
        (trim_data['매출액'] >= 1) & 
        (trim_data['PER'] >= 0) & 
        (trim_data['PBR'] <= 2)]
    
    trim_data['reciprocal_of_PBR'] = 1/trim_data['PBR']
    
    blue_chip = trim_data.sort_values(by='reciprocal_of_PBR', ascending=False ).head(100)
    return blue_chip

# 3. 우량주 중 오르고 있는 주식을 선정한다.
# 상장주식수의 1% 이상 거래량, 전일 대비 등락률 0%이상, 현재가가 시가보다 높은 주식 중에서 고가 근처에 있는 주식 10개를 선정.
def getTop10(blue_chip):
    blue_chip = blue_chip.loc[(blue_chip['거래량'] >= blue_chip['상장주식수']*1000/100) & (blue_chip['등락률'].str[0] != '-') & (blue_chip['현재가'] > blue_chip['시가'])]
    # 고가 근처에 있는 주식을 10개 선정
    # 근처 → 고가를 1로 가정했을 때 종가가 1에 가장 가까운 주식을 확인
    # ex) 주식1 : 고가2 종가1
    #     주식2 : 고가3 종가2
    #     2가 고가에 더 가깝다고 판정
    #     → 종가/고가가 높을수록 가깝다고 판정
    blue_chip = blue_chip.loc[blue_chip['현재가'] > blue_chip['시가']]
    blue_chip['closing_price_percentage'] = blue_chip['현재가']/blue_chip['고가']
    top10 = blue_chip.sort_values(by='closing_price_percentage', ascending=False).head(10)
    return top10



print("불필요한 컬럼을 제거")
trimData = getTrimData(base_stock_data)
print(trimData.columns)
print(trimData)
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
print("선정된 우량주식100")
blue_chip = getBlueChip100(trimData)
print(blue_chip)
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
print("우량주식 중 오르고 있는 주식 10개 선정")
top10 = getTop10(blue_chip)
print(top10)
```

```python
import pandas as pd


class StockAnalyzer:
    def __init__(self,path):
        self.data = pd.read_excel(path)

    def preprocess_data(self):
        # Remove unnecessary columns
        remove = ['액면가', '거래대금', '외국인비율', '매수호가', '매도호가']
        df_cleaned = self.data.drop(columns=remove)
        
        # Convert all numbers to float
        for col in df_cleaned.columns:
            if col != '종목명' and df_cleaned[col].dtype == 'O': 
                df_cleaned[col] = df_cleaned[col].str.replace(',', '').str.replace('%', '').astype(float)
        
        # Drop NaN rows
        self.data = df_cleaned.dropna(subset=['시가총액', 'PER', 'PBR', '매출액'])

    def filter_stocks(self):
        return self.data[
            (self.data['시가총액'] >= 1000) &
            (self.data['매출액'] >= 1) &
            (self.data['PER'] >= 0) &
            (self.data['PBR'] <= 2)
        ]

    def get_top_stocks(self, stocks):
        stocks['PBR_inverse'] = 1 / stocks['PBR']
        return stocks.sort_values(by='PBR_inverse', ascending=False).head(100)

    def get_potential_stocks(self, stocks):
        potential_stocks = stocks[
            (stocks['거래량'] >= stocks['상장주식수'] * 10) &
            (stocks['등락률'] >= 0) &
            (stocks['현재가'] > stocks['시가'])
        ]
        potential_stocks['고가-현재가'] = (potential_stocks['고가'] - potential_stocks['현재가'])
        return potential_stocks.sort_values(by='고가-현재가').head(10)

    def analyze(self):
        self.preprocess_data()
        
        filtered_stocks = self.filter_stocks()
        top_stocks = self.get_top_stocks(filtered_stocks)
        
        print('------ 문제 2 ------')
        print(top_stocks['종목명'].tolist())
        
        potential_stocks = self.get_potential_stocks(top_stocks)
        
        print("------ 문제 3 ------")
        print(potential_stocks['종목명'].tolist())

# Run the analyzer
analyzer = StockAnalyzer('data/NaverFinance.xlsx')
analyzer.analyze()
```
