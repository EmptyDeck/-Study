```python
import torch
import torch.optim as optim

import matplotlib.pyplot as plt
###how do you ouput oly colom row?
#cnt

#DATA
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]] #0 means False 1 means True

#cnt = 6
x_tensor = torch.FloatTensor(x_data)
y_tensor = torch.FloatTensor(y_data)


print(x_data)
print(x_tensor)
print(x_tensor.shape)
print(y_data)
print(y_tensor)
print(y_tensor.shape)
#SET W AND b
W = torch.zeros((2,1),requires_grad=True) #torch.ones makes a tensor full of '1' zeors makes full of 0's // we make 2,1 becuase we have two inputs and we need W sepretly
b = torch.zeros(1, requires_grad=True)

print(W)
print(W.shape)
print(b)
print(b.shape)

# Set Optimser as SGD (efficient one)
optimizer = optim.SGD([W,b], lr = 1)

nb_epochs = 1000 #epochs means THE moment

for epoch in range(nb_epochs +1 ):
  #HYPOTHESIS
  hypothesis = torch.sigmoid(x_tensor.matmul(W) + b )  #this is when W is a 2 dimensional matrix// i guess matmul is a short term of matrix muliply
  #COST
  cost = -(y_tensor * torch.log(hypothesis) + (1 - y_tensor) * torch.log(1 - hypothesis)).mean() #torch.log means log in math // (1-y) , y basicly means if 1 elif 0

  #IMPOVE MODEL
  optimizer.zero_grad() # reset gread to 0
  cost.backward()
  optimizer.step()
  #PRINT LOG
  if epoch % 100 == 0:
    x1_W = W[0][0]
    x2_W = W[1][0]
    print("No. %4d/%d W1 : %f W2 : %f b : %f Cost : %f "%(epoch,nb_epochs,x1_W.item(),x2_W.item(),b.item(),cost.item()))

```

```python
#TEST
print("hypothesis")
print(hypothesis)
prediction = hypothesis >= torch.FloatTensor([0.5])
print("prediction")
print(prediction)
correct_prediction = prediction.float() == y_tensor
print("Corrent_prediction")
print(correct_prediction)

```
