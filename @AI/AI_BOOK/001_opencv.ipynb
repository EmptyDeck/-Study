```python
import cv2 as cv
import sys

# 카메라와 연결 시도
try:
    cap = cv.VideoCapture(0, cv.CAP_DSHOW)  # 카메라 연결 시도 (Windows용)
except:
    cap = cv.VideoCapture(0)
if not cap.isOpened():
    sys.exit('카메라 연결 실패')

while True:
    ret, frame = cap.read()  # 비디오 프레임 획득

    # 프레임 획득에 실패한 경우 루프 종료
    if not ret:
        print('프레임 획득에 실패하여 루프를 나갑니다.')
        break

    cv.imshow('Video display', frame)

    key = cv.waitKey(1)

    # 'q' 키를 누르면 루프 종료
    if key == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('a.jpg')

grayimg = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

plt.imshow(grayimg,cmap='gray')
plt.show()
```

```python
h = cv.calcHist([grayimg],[0],None,[256],[0,256])
plt.plot(h,color = 'r' , linewidth = 1)
plt.show()
```

```python
equal = cv.equalizeHist(grayimg)
plt.imshow(equal,cmap='gray')
plt.show()
```
